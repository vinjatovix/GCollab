```python
import random
from os import system, name

def clear():
    if name == 'nt': 
        _ = system('cls')
    else: 
        _ = system('clear')

lista = ['ajo','abanico','bateria']
objetivo = random.choice(lista)


letras = ''

intentos = 1
aciertos = 0
incognita = list('*' * len(objetivo))

def reset():
    global intentos
    intentos = 1
    global aciertos
    aciertos = 0
    global objetivo
    objetivo = random.choice(lista) 
    global incognita
    incognita = list('*' * len(objetivo))

reset()

def restart():
    again = input('¿Quieres jugar de nuevo? s/n :')
        
    if again != 's':
        print('\nGracias por jugar.\n... ... ... Saliendo.')
    else:
        clear()
        print('\n### INICIANDO UNA NUEVA PARTIDA ###')
        reset()
        status()
    

# muñeco
p5   = '________'
p4   = '|/  |'
p42  = '|/'
p3   = '|   0'
p32  = '|'
p2   = '|  /|\\'
p23  = '|  /'
p1   = '|  / \\'
base = '|_______'

def muerto():
    print(p5)
    print(p4)
    print(p3)
    print(p2)
    print(p1)
    print(base)
    
def menosUno():
    print(p5)
    print(p42)
    print(p3)
    print(p2)
    print(p1)
    print(base)
    
def menosDos():
    print(p5)
    print(p42)
    print(p32)
    print(p2)
    print(p1)
    print(base)
    
def menosTres():
    print(p5)
    print(p42)
    print(p32)
    print(p1)
    print(p1)
    print(base)
    
def menosCuatro():
    print(p5)
    print(p42)
    print(p32)
    print(p23)
    print(p1)
    print(base)
    
def menosCinco():
    print(p5)
    print(p42)
    print(p32)
    print(p32)
    print(p1)
    print(base)

def menosSeis():
    print(p5)
    print(p42)
    print(p32)
    print(p32)
    print(p23)
    print(base)
    
def menosSiete():
    print(p5)
    print(p42)
    print(p32)
    print(p32)
    print(p32)
    print(base)
    
def menosOcho():
    print(' ')
    print(p42)
    print(p32)
    print(p32)
    print(p32)
    print(base)

def menosNueve():
    print(' ')
    print(p32)
    print(p32)
    print(p32)
    print(p32)
    print(base)

def inicio():
    print('\n\n\n\n\n','_'+base[1:])

def banner():
    print('Llevas '+str(aciertos)+' aciertos')
    print('Te quedan '+str(intentos)+' intentos.')
    print('Adivina la palabra: ',incognita)
    
    if intentos == 10:
        inicio()
    elif intentos == 9:
        menosNueve()
    elif intentos == 8:
        menosOcho()
    elif intentos == 7:
        menosSiete()
    elif intentos == 6:
        menosSeis()
    elif intentos == 5:
        menosCinco()
    elif intentos == 4:
        menosCuatro()
    elif intentos == 3:
        menosTres()
    elif intentos == 2:
        menosDos()
    elif intentos == 1:
        menosUno()
    else:
        muerto()
  
```

```python
def status():
    banner()
    if aciertos >= len(objetivo):
        
        print('HAS GANADO\n')
        restart()
     
    elif intentos <= 0:
        print('HAS MUERTO\n')
        restart()

    else:
        jugar()
```

```python
def jugar():
    global intentos
    global aciertos
    paso = 0
    i = 0
    letraValida = 0
    while letraValida == 0:
        prueba = input('\nescoge una letra: ').lower()
    
        try:
            int(prueba)
            print('Eso no es una letra')
        except:
            print('\nA ver que ocurre...')
        
        if len(prueba) > 1:
            print('Eso son varias letras... Por favor introduce UNA letra')
        elif len(prueba) < 1:
            print('Introduce por lo menos UNA letra')
        else:
            letraValida = 1
            
    for letra in objetivo:
        if letra == prueba:
            print('\nHas acertado la letra',letra)
            del incognita[i]
            incognita.insert(i,prueba)
            i += 1
            aciertos += 1
            paso += 1

    if paso == 0:
        print('pierdes un intento')
        intentos -= 1
        
    status()
```
