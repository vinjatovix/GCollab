# Matplotlib
### es una libreria de python que nos permite representar datos de forma gr√°fica

```python
import matplotlib.pyplot as plt
```

```python
# vamos a apoyarnos en numpy para crear los datos rapidamente

import numpy as np
import pandas as pd

x = np.linspace(0,10,20)
y = x * 2
z = x * x
```

```python
# dibujamos la gr√°fica
plt.plot(x,y)

# etiquetamos la gr√°fica
plt.title('TEST')
plt.xlabel('eje de las x')
plt.ylabel('eje de las y')
```

### podemos distribuir varias gr√°ficas dentro del mismo espacio

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,3)
plt.plot(x,z)
plt.subplot(3,2,4)
plt.plot(y,y)
plt.subplot(3,2,5)
plt.plot(y,x)
plt.subplot(3,2,6)
plt.plot(z,y)
plt.xlabel('eje X')
plt.tight_layout()
```

```python
plt.scatter(x,z)
```

```python
plt.bar(x,z)
```

```python
plt.hist(x,z)
```

```python
plt.fill(x,y)
```

```python
plt.polar(x,x)
```

```python
plt.plot(x,y,label = 'ejemplo')
plt.legend(loc = 7)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('eje X')
axes.set_ylabel('eje y')
axes.set_title('prueba')
axes.plot(x,y)
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1,0.1,1,1])
x1.plot(x,y,label = 'ejemplo')
x1.plot(x,x,label = 'E2')
plt.legend(loc = 7)
x1.set_xlabel('eje X')
x1.set_ylabel('eje Y')
x1.set_title('grande')

x2 = fig.add_axes([1.2,0.8,0.3,0.3])
x2.plot(z,x, label = 'ex')
plt.legend(loc = 7)
x2.set_xlabel('eje X')
x2.set_ylabel('eje Y')
x2.set_title('pq')

```

```python
fig , axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[1].plot(x,x)


plt.tight_layout()
```

```python
fig , axes = plt.subplots(2,2)

axes[(1,0)].plot(x,y)
axes[(0,1)].plot(x,x)


plt.tight_layout()
```

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,label = 'ejemplo')
plt.legend(loc = 7)
```

```python
fig , axes = plt.subplots(2,1, figsize=(5,5))
axes[0].plot(x,y, label='test')
axes[0].legend(loc = 7)
axes[0].set_xlabel('eje X')
axes[0].set_ylabel('eje Y')
axes[0].set_title('arriba')

axes[1].plot(y,x)
axes[1].set_xlabel('eje X')
axes[1].set_ylabel('eje Y')
axes[1].set_title('abajo')

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,label = 'ejemplo',color ='r', linewidth=10, alpha = 0.5, marker = 'o',markersize=20,markerfacecolor = 'green')
axes.plot(x,x,label = 'e2',color='orange', lw=5, linestyle = '--')
plt.legend(loc = 7)
```

```python
# Limitando la grafica


fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('eje X')
axes.set_ylabel('eje y')
axes.set_title('prueba')
axes.plot(x,y)
axes.set_ylim([0,60])
```
